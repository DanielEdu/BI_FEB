Matplotlib es una biblioteca ampliamente utilizada en Python para la generación de gráficos y visualización de datos. 
Proporciona una amplia variedad de herramientas y funciones para crear visualizaciones de alta calidad

Antes de usar Matplotlib, asegúrate de que esté instalado en tu entorno de Python. 
Puedes instalarlo utilizando el administrador de paquetes pip con el siguiente comando:
pip install matplotlib


Para utilizar Matplotlib, debes importar el módulo correspondiente. A menudo, se importa bajo el alias plt para mayor comodidad:
`import matplotlib.pyplot as plt`

```python
import matplotlib.pyplot as plt
```

```python
# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
```

```python
# Crear el gráfico
plt.plot(x, y)
# Personalizar el gráfico
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de línea')

# Mostrar el gráfico
plt.show()
```

```python
# Datos de ejemplo
categorias = ['A', 'B', 'C', 'D']
valores = [20, 35, 30, 25]
```

```python
# Crear el gráfico de barras
plt.bar(categorias, valores)
# Personalizar el gráfico
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.title('Gráfico de barras')

# Mostrar el gráfico
plt.show()
```
